### id()    - used to verify the identity of any variable
### type()  - used to identify the type of the variable

````python
print(type(a))
c = 'a'
print(type(c))
c = 122342342222234.3
print(type(c))
````

````python
c = 123
print(id(c))
c = 12.3
print(id(c))
df = 123
print(id(df))
df = "ad"
print(id(df))
````

````python
t = "ad"
print(id(t))
````
